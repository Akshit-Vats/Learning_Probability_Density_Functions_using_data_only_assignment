```python

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Step 1: Load data
data = pd.read_csv("data.csv", encoding_errors="ignore")  
x = data["no2"].dropna().values

# roll number
r = 102317091

a_r = 0.5 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)

z = x + a_r * np.sin(b_r * x)
z = torch.tensor(z, dtype=torch.float32).view(-1, 1)

# Step 2: Models
latent_dim = 5

class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(latent_dim, 16),
            nn.ReLU(),
            nn.Linear(16, 16),
            nn.ReLU(),
            nn.Linear(16, 1)
        )
    def forward(self, x):
        return self.net(x)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 16),
            nn.ReLU(),
            nn.Linear(16, 16),
            nn.ReLU(),
            nn.Linear(16, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.net(x)

G = Generator()
D = Discriminator()

# Step 3: Training
criterion = nn.BCELoss()
g_opt = optim.Adam(G.parameters(), lr=0.001)
d_opt = optim.Adam(D.parameters(), lr=0.001)

epochs = 2000
batch_size = 64

for epoch in range(epochs):
    idx = np.random.randint(0, len(z), batch_size)
    real = z[idx]

    real_labels = torch.ones(batch_size, 1)
    fake_labels = torch.zeros(batch_size, 1)

    # Train Discriminator
    noise = torch.randn(batch_size, latent_dim)
    fake = G(noise)

    d_loss = criterion(D(real), real_labels) + criterion(D(fake.detach()), fake_labels)
    d_opt.zero_grad()
    d_loss.backward()
    d_opt.step()

    # Train Generator
    noise = torch.randn(batch_size, latent_dim)
    fake = G(noise)

    g_loss = criterion(D(fake), real_labels)
    g_opt.zero_grad()
    g_loss.backward()
    g_opt.step()

# Step 4: Generate samples
noise = torch.randn(50000,latent_dim )
gen_samples = G(noise).detach().numpy()

# Step 5: KDE for PDF
kde = KernelDensity(kernel='gaussian', bandwidth=1.0).fit(gen_samples)

z_range = np.linspace(gen_samples.min(), gen_samples.max(), 500).reshape(-1, 1)
log_density = kde.score_samples(z_range)

plt.plot(z_range, np.exp(log_density))
plt.title("Estimated PDF using GAN")
plt.show()
```
